```python
#For missing values find and replace with arithmetic mean of the attributes
#Find and replace outliers with median
#Compare the 5 number summary before and after preprocessing
df2=df.replace(-1,np.nan)
```

```python
df_cleaned=df2.fillna(df.mean())
```

```python
df_cleaned.head()
```

```python
#Find and replace outliers with median
columns=['calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups',
       'rating']
for col in columns:
    describe=df_cleaned[col].describe()
    q3=describe['75%']
    q1=describe['25%']
    iqr=q3-q1
    upper_bound=q3+1.5*iqr
    lower_bound=q1-1.5*iqr
    for i in df_cleaned[col]:
        if(i>upper_bound or i<lower_bound):
            df_cleaned=df_cleaned.replace(i,df.median())
```
